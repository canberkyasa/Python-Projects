Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")
# Preview the data
schools.head()
```

```python
# Finding best math results by school

best_math_schools = schools[schools["average_math"] > (0.8 * 800)]
best_math_schools = best_math_schools.loc[:,["school_name","average_math"]].sort_values(by="average_math",ascending=False)
best_math_schools
```

```python
# Finding top 10 performing schools based on combined SAT score
total_SAT = schools["average_math"] + (schools["average_reading"] ) + (schools["average_writing"])
top_10_schools = pd.concat([schools[["school_name"]],total_SAT.rename("total_SAT")],axis=1)
top_10_schools = top_10_schools.sort_values(by="total_SAT",ascending=False)
top_10_schools = top_10_schools.iloc[:10]
top_10_schools.head(10)
```

```python
# Adding total SAT score to schools df
schools_total_SAT = pd.concat([schools,total_SAT.rename("total_SAT")], axis=1)
schools_total_SAT.head()
```

```python
# Groupping by boroughs and finding the largest standard deviation in the combined SAT score.
boroughs = schools_total_SAT.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)

largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev

largest_std_dev.reset_index(inplace=True)
largest_std_dev
```
